```python
import os
import matplotlib.pyplot as plt

# Directory containing the files
dataset_path = "audio_representations/audio_representations_ravdess/spectrograms/"

# Emotion mapping
emotion_map = {
    "01": "Neutral",
    "02": "Calm",
    "03": "Happy",
    "04": "Sad",
    "05": "Angry",
    "06": "Fearful",
    "07": "Disgust",
    "08": "Surprised",
}

# Initialize a dictionary to count frequencies
emotion_counts = {emotion: 0 for emotion in emotion_map.values()}

# Iterate through the files in the directory
for filename in os.listdir(dataset_path):
    if filename.endswith(".png"):
        emotion_code = filename.split("-")[2]  # Extract the 3rd part of the name
        emotion = emotion_map.get(emotion_code, "Unknown")
        if emotion in emotion_counts:
            emotion_counts[emotion] += 1

# Create the bar plot
plt.figure(figsize=(10, 6))  # Larger figure for better visibility
plt.bar(emotion_counts.keys(), emotion_counts.values(), color="#1f77b4", edgecolor="black")

# Add titles and labels in English
plt.title("Emotion Distribution in the RAVDESS Dataset", fontsize=16, fontweight="bold")
plt.xlabel("Emotions", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Display the plot
plt.show()
```
