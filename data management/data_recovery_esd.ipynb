# Normalisation

```python
# Load file
file_path = "Emotion Speech Dataset/0011/Angry/0011_000351.wav"
y, sr = librosa.load(file_path, sr=None)  # Load with native sample rate

print("Sample rate : " + str(sr))

# Play original audio
print("Original audio :")
display(Audio(y, rate=sr))
# Show original sound wave
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Original sound wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

# Normalisation bas√©e sur le maximum d'amplitude
y_normalized = y / np.max(np.abs(y))

# Play normalized audio
print("Audio after normalization :")
display(Audio(y_normalized, rate=sr))

# Show normalized sound wave
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_normalized, sr=sr)
plt.title("Normalized sound wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
